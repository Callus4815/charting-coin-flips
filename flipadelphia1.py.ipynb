```python
import random
import matplotlib.pyplot as plt
import statistics
```

```python

%matplotlib inline

def coin_flip():
    
    for _ in  range(1) :
        coin = random.randint(1,2)
        if coin == 1:
            return 'H'     
        else:            
            return 'T'
        
```

```python
def simulator(x=16):
    count = 0
    head = []
    tail = []
    count_head = 0
    count_tail = 0
    
    while count < 2 ** x:
        coin = coin_flip()
        if coin == 'H':
            count_head+=1
        else:
            count_tail+=1
        count+=1
        for i in range(x + 1):
            if count == 2**i: 
                head.append(count_head)
                tail.append(count_tail)
    return head,  tail

head,tail = simulator()
```

```python
diff = [head[i]-tail[i] for i in range(len(head))]
        
        
```

```python
print(diff)
```

```python
x = diff
y = 17
plt.plot(x)

plt.xticks(range(17))
plt.xlabel("Number of flips")
plt.ylabel("Heads or Tails")
plt.title("Differences of Flips")

plt.show()
```
